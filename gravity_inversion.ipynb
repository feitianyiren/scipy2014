```python
import numpy as np
from fatiando.gravmag import talwani
from fatiando.mesher import Polygon
from fatiando.inversion.base import Misfit

class BasinPolygonal(Misfit):
    def __init__(self, x, z, gz, height, npoints, density):
        super(BasinPolygonal, self).__init__(
            data=gz, 
            positional=dict(x=x, z=z),
            model=dict(height=height, props={'density': density}),
            nparams=npoints, islinear=False)
    
    def p2vertices(self, p):
        n = self.nparams + 2
        vertices = np.empty((n, 2))
        x = self.positional['x']
        vertices[:, 0] = np.linspace(x.min(), x.max(), n)[::-1]
        h = self.model['height']
        vertices[:, 1] = np.concatenate([[h], p, [h]])
        return vertices
    
    def _get_predicted(self, p):
        x, z = self.positional['x'], self.positional['z']
        poly = Polygon(self.p2vertices(p), self.model['props'])
        return talwani.gz(x, z, [poly])
        
    def _get_jacobian(self, p):    
        x, z = self.positional['x'], self.positional['z']
        verts = self.p2vertices(p)
        props = self.model['props']
        dz = np.array([0, 10])
        jac = np.zeros((self.ndata, self.nparams))
        for i in xrange(self.nparams):
            diff = Polygon([verts[i + 2], verts[i + 1] - dz, 
                            verts[i], verts[i + 1] + dz], props)
            jac[:, i] = talwani.gz(x, z, [diff])/(2 * dz[1])
        return jac
    
    def fit(self):
        super(BasinPolygonal, self).fit()
        self._estimate = Polygon(self.p2vertices(self.p_), 
                                 self.model['props'])
        return self
```

```python
def confidence(obj):
    dof = obj.ndata - obj.nparams
    variance = np.linalg.norm(obj.residuals(obj.p_))**2/dof
    cov = variance*np.linalg.inv(obj.hessian(obj.p_))
    std = np.sqrt(np.diagonal(cov))
    verts = np.concatenate([obj.p2vertices(obj.p_ - 2*std),
                            obj.p2vertices(obj.p_ + 2*std)[1:-1][::-1]])
    return Polygon(verts)
```

# Test on some synthetics

```python
x = np.linspace(0, 100000, 100)
z = -100*np.ones_like(x)
xs = np.linspace(0, 100000, 100)[::-1]
depths = (-1e-15*(xs - 50000)**4 + 8000 
          - 3000*np.exp(-(xs - 70000)**2/(10000**2)))
depths -= depths.min()
density = -500
model = Polygon(np.transpose([xs, depths]), {'density':density})
gz = utils.contaminate(talwani.gz(x, z, [model]), 0.1, seed=0)
```

```python
mpl.figure(figsize=(10, 6))
mpl.subplot(2, 1, 1)
mpl.plot(x*0.001, gz, '.k')
mpl.xlabel('x (km)')
mpl.ylabel('Gravity anomaly (mGal)')
mpl.ylim(-130, 0)
mpl.subplot(2, 1, 2)
mpl.polygon(model, fill='grey')
mpl.ylim(0, 10000)
mpl.m2km()
ax = mpl.gca()
ax.invert_yaxis()
ax.set_xticklabels([''])
mpl.ylabel('depth (km)')
mpl.text(20000, 3000, 'Sediments')
mpl.text(15000, 8000, 'Basement')
```

```python
misfit = BasinPolygonal(x, z, gz, 0, 30, density)
initial = 3000*np.ones(misfit.nparams)
misfit.config('levmarq', initial=initial).fit()
```

```python
mpl.figure(figsize=(10, 6))
mpl.subplot(2, 1, 1)
mpl.plot(x*0.001, gz, 'ok')
mpl.plot(x*0.001, misfit.predicted(), '-r')
mpl.ylim(-130, 0)
mpl.xlabel('x (km)')
mpl.ylabel('Gravity anomaly (mGal)')
mpl.subplot(2, 1, 2)
mpl.polygon(model, fill='grey')
mpl.polygon(confidence(misfit), '-b', fill='blue', alpha=0.7)
mpl.polygon(misfit.estimate_, 'o-r')
mpl.ylim(0, 10000)
mpl.m2km()
ax = mpl.gca()
ax.invert_yaxis()
ax.set_xticklabels([''])
mpl.ylabel('depth (km)')
```

```python
from fatiando.inversion.regularization import Smoothness1D
solver = misfit + 1e-9*Smoothness1D(misfit.nparams)
solver.config('levmarq', initial=initial).fit()
```

```python
mpl.figure(figsize=(10, 6))
mpl.subplot(2, 1, 1)
mpl.plot(x*0.001, gz, 'ok')
mpl.plot(x*0.001, solver.predicted(), '-r')
mpl.xlabel('x (km)')
mpl.ylabel('Gravity anomaly (mGal)')
mpl.ylim(-130, 0)
mpl.subplot(2, 1, 2)
mpl.polygon(model, fill='grey')
mpl.polygon(confidence(solver), '-b', fill='blue', alpha=0.7)
mpl.polygon(solver.estimate_, 'o-r')
mpl.ylim(0, 10000)
mpl.m2km()
ax = mpl.gca()
ax.invert_yaxis()
ax.set_xticklabels([''])
mpl.ylabel('depth (km)')
```

```python
from fatiando.inversion.regularization import LCurve
lc = LCurve(misfit, Smoothness1D(misfit.nparams),
            [10**i for i in xrange(-15, -5)], jobs=4)
lc.config('levmarq', initial=initial).fit()
```

```python
print(lc.regul_param_)
lc.plot_lcurve()
mpl.set_area([5e-3, 1, 5e6, 5e7])
```

```python
mpl.figure(figsize=(10, 6))
mpl.subplot(2, 1, 1)
mpl.plot(x*0.001, gz, 'ok')
mpl.plot(x*0.001, lc.predicted(), '-r')
mpl.xlabel('x (km)')
mpl.ylabel('Gravity anomaly (mGal)')
mpl.ylim(-130, 0)
mpl.subplot(2, 1, 2)
mpl.polygon(model, fill='grey')
mpl.polygon(confidence(lc.objectives[lc.corner_]), '-b', fill='blue', alpha=0.7)
mpl.polygon(lc.estimate_, 'o-r')
mpl.ylim(0, 10000)
mpl.m2km()
ax = mpl.gca()
ax.invert_yaxis()
ax.set_xticklabels([''])
mpl.ylabel('depth (km)')
```
