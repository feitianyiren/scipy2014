```python
import numpy as np
from fatiando.gravmag import talwani
from fatiando.mesher import Polygon
from fatiando.inversion.base import Misfit

class BasinPolygonal(Misfit):
    def __init__(self, x, z, gz, height, npoints, density):
        super(BasinPolygonal, self).__init__(
            data=gz, 
            positional=dict(x=x, z=z),
            model=dict(height=height, props={'density':density}),
            nparams=npoints, islinear=False)
    
    def p2vertices(self, p):
        n = self.nparams + 2
        vertices = np.empty((n, 2))
        x = self.positional['x']
        vertices[:,0] = np.linspace(x.min(), x.max(), n)[::-1]
        h = self.model['height']
        vertices[:,1] = np.concatenate([[h], p, [h]])
        return vertices
    
    def _get_predicted(self, p):
        x, z = self.positional['x'], self.positional['z']
        poly = Polygon(self.p2vertices(p), self.model['props'])
        return talwani.gz(x, z, [poly])
        
    def _get_jacobian(self, p):    
        x, z = self.positional['x'], self.positional['z']
        verts = self.p2vertices(p)
        props = self.model['props']
        dz = np.array([0, 10])
        jac = np.zeros((self.ndata, self.nparams))
        for i in xrange(1, self.nparams + 1):
            diff = Polygon([verts[i + 1], verts[i] - dz, 
                            verts[i - 1], verts[i] + dz], props)
            jac[:, i - 1] = talwani.gz(x, z, [diff])/(2*dz[1])
        return jac
    
    def fit(self):
        super(BasinPolygonal, self).fit()
        self._estimate = Polygon(self.p2vertices(self.p_), 
                                 self.model['props'])
        return self    
```

# Test on some synthetics

```python
from fatiando import utils
```

```python
x = np.linspace(0, 100000, 100)
z = -100*np.ones_like(x)
xs = np.linspace(0, 100000, 100)[::-1]
depths = (-1e-15*(xs - 50000)**4 + 6000 
          - 3000*np.exp(-(xs - 70000)**2/(10000**2)))
depths -= depths.min()
density = -500
model = Polygon(np.transpose([xs, depths]), {'density':density})
gz = utils.contaminate(talwani.gz(x, z, [model]), 0.1, seed=0)
```

```python
mpl.figure(figsize=(10, 6))
mpl.subplot(2, 1, 1)
mpl.plot(x*0.001, gz, 'ok')
mpl.xlabel('x (km)')
mpl.ylabel('Gravity anomaly (mGal)')
mpl.subplot(2, 1, 2)
mpl.polygon(model, fill='grey')
mpl.m2km()
ax = mpl.gca()
ax.invert_yaxis()
ax.set_xticklabels([''])
mpl.ylabel('depth (km)')
```

```python
misfit = BasinPolygonal(x, z, gz, 0, 100, density)
initial = 3000*np.ones(misfit.nparams)
misfit.config('levmarq', initial=initial).fit()
```

```python
mpl.figure(figsize=(10, 6))
mpl.subplot(2, 1, 1)
mpl.plot(x*0.001, gz, 'ok')
mpl.plot(x*0.001, misfit.predicted(), '-r')
mpl.xlabel('x (km)')
mpl.ylabel('Gravity anomaly (mGal)')
mpl.subplot(2, 1, 2)
mpl.polygon(model, fill='grey')
mpl.polygon(misfit.estimate_, 'o-r')
mpl.m2km()
ax = mpl.gca()
ax.invert_yaxis()
ax.set_xticklabels([''])
mpl.ylabel('depth (km)')
```

```python
from fatiando.inversion.regularization import Smoothness1D
solver = misfit + 1e-8*Smoothness1D(misfit.nparams)
solver.config('levmarq', initial=initial).fit()
```

```python
mpl.figure(figsize=(10, 6))
mpl.subplot(2, 1, 1)
mpl.plot(x*0.001, gz, 'ok')
mpl.plot(x*0.001, solver.predicted(), '-r')
mpl.xlabel('x (km)')
mpl.ylabel('Gravity anomaly (mGal)')
mpl.subplot(2, 1, 2)
mpl.polygon(model, fill='grey')
mpl.polygon(solver.estimate_, 'o-r')
mpl.m2km()
ax = mpl.gca()
ax.invert_yaxis()
ax.set_xticklabels([''])
mpl.ylabel('depth (km)')
```

```python
from fatiando.inversion.regularization import LCurve
lc = LCurve(misfit, Smoothness1D(misfit.nparams),
            [10**i for i in xrange(-15, -5)], jobs=4)
lc.config('levmarq', initial=initial).fit()
```

```python
lc.plot_lcurve()
```

```python
mpl.figure(figsize=(10, 6))
mpl.subplot(2, 1, 1)
mpl.plot(x*0.001, gz, 'ok')
mpl.plot(x*0.001, lc.predicted(), '-r')
mpl.xlabel('x (km)')
mpl.ylabel('Gravity anomaly (mGal)')
mpl.subplot(2, 1, 2)
mpl.polygon(model, fill='grey')
mpl.polygon(lc.estimate_, 'o-r')
mpl.m2km()
ax = mpl.gca()
ax.invert_yaxis()
ax.set_xticklabels([''])
mpl.ylabel('depth (km)')
```
